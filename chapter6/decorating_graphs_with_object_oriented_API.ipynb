```python
# 출판사 별 발행 도서 개수 산점도 그리기
# 데이터 프레임 다운 및 읽기
import gdown
gdown.download('http://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)

import pandas as pd
ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
# 고유한 출판사 목록 만들기
# 상위 30개 출판사 선택
top30_pubs = ns_book7['출판사'].value_counts()[:30]
top30_pubs
```

```python
# 상위 30개 출판사에 해당되는 지 검사
top30_pubs_idx = ns_book7['출판사'].isin(top30_pubs.index)
top30_pubs_idx
```

```python
# 데이터 프레임의 행 1000개를 무작위로 선택
ns_book8 = ns_book7[top30_pubs_idx].sample(1000, random_state=42)
ns_book8.head()
```

```python
# 산점도 그리기
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(ns_book8['발행년도'], ns_book8['출판사'])
ax.set_title('출판사별 발행 도서')
fig.show()
```

```python
# 값에 따라 마커 크기를 다르게 나타내기
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(ns_book8['발행년도'], ns_book8['출판사'], s=ns_book8['대출건수'])
ax.set_title('출판사별 발행 도서')
fig.show()
```

```python
# 마커 꾸미기
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(ns_book8['발행년도'], ns_book8['출판사'], s = ns_book8['대출건수']*2, alpha=0.3, edgecolors='k', linewidths=0.5, c=ns_book8['대출건수'])
ax.set_title('출판사별 발행 도서')
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sc = ax.scatter(ns_book8['발행년도'], ns_book8['출판사'], alpha=0.3, s=ns_book8['대출건수']*2, edgecolors='k', linewidths=0.5, c=ns_book8['대출건수'], cmap='jet')
ax.set_title('출판사별 발행 도서')
fig.colorbar(sc)
fig.show()
```
