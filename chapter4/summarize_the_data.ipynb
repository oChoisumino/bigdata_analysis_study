# 4. 데이터 요약하기

## 4-1. 통계로 요약하기
* 기술통계 구하기
  - 기술 통계 (= 요약통계)
    * 자료의 내용을 압축하여 설명하는 방법
    * 평균, 표준편차 등이 있다.
  - describe() 메서드
    * 기본적인 몇 가지 기술통계를 자동으로 추출
      - count : 누락된 값을 제외한 데이터의 개수
      - mean : 평균 
      - std : 표준편차
      - min : 최솟값
      - 50% : 중앙값
      - 25%, 75% : 순서대로 늘어놓았을 때, 25% 지점과 75% 지점에 놓인 값
      - max : 최댓값
    * 매개변수
      - percentiles : 위치를 지정하면 원하는 위치의 값을 볼 수 있다. ([ ]로 전달 가능)
      - include : 열의 데이터 타입이 수치가 아닌 다른 데이터 타입의 기술통계를 보고싶을 때 데이터 타입 지정
        * object로 지정할 경우 count, unique, top, freq 등의 통계가 나온다.
        * count : 누락된 값을 제외한 데이터의 개수 
        * unique : 고유한 값의 개수를 나타낸다.
        * top : 가장 많이 등장하는 값
        * freq : top에 등장하는 항목의 빈도

* 평균 구하기
  - 숫자의 값을 모두 더해 개수로 나눈 값
  - $$y=\frac{\sum_{k=1}^n x_k}{n}$$
  - mean() 메서드 사용

* 중앙값 구하기
  - 전체 데이터를 순서대로 늘어 놓았을 때 중앙에 위치한 값
  - 데이터 개수가 짝수인 경우 가운데 두 개의 값을 평균하여 중앙값을 결정
  - median() 메서드 사용

* 최솟값 구하기
  - 전체 데이터 중 가장 작은 값
  - min() 메서드 사용

* 최댓값 구하기
  - 전체 데이터 중 가장 큰 값
  - max() 메서드 사용

* 분위수 구하기
  - 데이터를 순서대로 늘어 놓았을 때 이를 균등한 간격으로 나누는 기준점
  - 사분위수 : 순서대로 정렬된 데이터를 네 구간으로 나누는 값 (0.25, 0.5, 0.75)
  - quantile() 메서드 사용
    * 매개변수 interpolation 
      - 중간 값을 계산하는 방법을 결정
      - midpoint : 분위수에 상관없이 무조건 두 수 사이의 중앙값을 사용
      - nearest : 두 수 중에서 가까운 값을 선택
      - lower : 두 수 중 작은 값을 선택
      - higher : 두 수 중 큰 값을 선택

* 분산 구하기
  - 평균으로부터 데이터가 얼마나 퍼져있는지를 나타내는 통계량
  - $$y = \frac{\sum_{i=1}^n (x_i - x)^2}{n}$$
    $$(n : 전체 샘플 개수, x : 평균)$$
  - var() 메서드 사용
    * 매개변수 ddof를 이용하여 자유도 차감값을 지정 가능

* 표준편차 구하기
  - 분산에 제곱근을 한 값
  - $$y = \sqrt \frac{\sum_{i=1}^n (x_i - x)^2}{n}$$
    $$(n : 전체 샘플 개수, x : 평균)$$
  - std() 메서드 사용

* 최빈값 구하기
  - 데이터에서 가장 많이 등장하는 값
  - mode() 메서드 사용

* 데이터프레임에서 기술통계 구하기
  - numeric_only 매개변수를 사용하여 수치형 데이터에만 적용할 수 있다.

* 넘파이의 기술통계 함수
  - 평균 : mean(), average() 함수 사용
  - 중앙값 : median() 함수 사용
  - 최솟값 : min() 함수 사용
  - 최댓값 : max() 함수 사용
  
  - 분위수 : quantile() 함수, percentile() 함수 사용
  - 분산 : var() 함수 사용
    * 표본집단 : 전체 데이터 중에서 수집한 일부 데이터
    * 모집단 : 전체 데이터
    * ddof 매개변수를 사용하여 자유도 차감값 지정 가능
  - 표준편차 : std() 함수 사용
  - 최빈값 : unique() 함수를 사용해 고유한 값과 등장 횟수 배열을 얻어, argmax로 가장 큰 값의 인덱스를 찾고 해당 인덱스의 값을 출력한다.



## 4-2. 분포 요약하기
* 산점도 그리기
  - 데이터를 화면에 뿌리듯 그리는 그래프
  - 두 특성(변수) 값을 직교 좌표계에 점으로 나타내는 그래프
  - matplotlib.pyplot 패키지 아래에 존재
  - scatter() 함수
    * 산점도를 그리는 함수
    * 예) plt.scatter(x좌표 리스트, y좌표 리스트)
    * 매개변수 alpha
      - 0~1 사이의 값으로 투명도를 지정
      - 값이 작을수록 투명해진다.
      - 데이터가 중첩된 부분은 짙게 표현
  - show() 함수
    * 그래프를 출력할 때 사용

* 히스토그램 그리기
  - 수치형 특성의 값을 일정한 구간으로 나누어 구간 안에 포함된 데이터 개수를 그래프로 그린 것
  - 구간 안에 속한 데이터 개수를 도수라고 한다.
  - matplotlib.pyplot 패키지 아래에 존재
  - hist() 함수
    * 히스토그램을 그리는 함수
    * 매개변수 
      - bins : 나누는 구간 개수
      - log : 로그 스케일로 히스토그램 그리기(True, False)
  - histogram_bin_edges() 함수
    * 구간의 경계 값을 얻는 함수
  - yscale() 함수
    * y축의 구간을 변경 가능
  - xscale() 함수
    * x축의 구간을 변경 가능

* 상자 수염 그리기
  - 방법
    * 1. 사분위수를 계산한다. 25%와 75% 지점을 밑면과 윗면으로 하는 직사각형을 그린다.
    * 2. 중간값, 즉 50%에 해당하는 지점에 수평선을 긋는다.
    * 3. 사각형의 밑면과 윗면에서 사각형의 높이의 1.5배 만큼 떨어진 거리안에서 가장 멀리 있는 샘플까지 수직선을 긋는다.
    * 4. 이 수직선 밖에서 최솟값과 최댓값까지 데이터를 점으로 표시한다. 이 영역의 데이터를 이상치라고 한다.
  - boxplot() 함수
    * 상자 수염 그림을 그리는 함수
    * 매개변수
      - vert : 수직으로 그리는 것을 정하는 매개변수(True, False)
      - whis  수염의 길이를 조정하는 매개변수로
        * default는 1.5(IQR의 1.5배)
        * 백분율로도 지정가능 -> (10,90)이면 10, 90 분위수에 해당하는 데이터까지 수염을 그린다.
