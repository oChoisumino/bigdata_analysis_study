# 7. 검증하고 예측하기

## 7-1. 통계적으로 추론하기
* 모수 검정
  - 모집단에 대한 파라미터를 추정하는 방법
    * 모집단 : 관심 대상이 되는 전체 데이터
    * 파라미터 : 평균, 분산 등 통계치
  - 표본 : 모집단에서 선택한 일부 샘플
  - 모수 검정은 데이터가 정규분포를 따른다는 가정을 전제로 함
    * 정규 분포 : 평균을 중심으로 대칭인 분포

* 표준 점수 (Z 점수)
  - 데이터가 정규분포를 따른다고 가정하고, 각 값이 평균에서 얼마나 떨어져 있는지 표준편차를 사용해 변환한 점수
  - Z점수는 평균까지 거리를 표준편차로 나눈 것
  - $$y=\frac{x-\mu}{\sigma}$$
  - zscore()
    * scipy 의 stats 모듈 내에 존재
    * z 점수를 구할 때 사용

* 누적분포
  - 표준정규분포 : 평균이 0이고 표준편차가 1인 정규분포
  - 표준 정규분포에서 z점수가 1 이내에 위치한 샘플은 전체의 약 68%에 해당
  - norm.cdf() 메서드
    * scipy 의 stats 모듈 내에 존재
    * 누적된 분포를 반환
    * cdf : 누적 분포 함수 (cumulative distribution function)
  - norm.ppf() 메서드
    * scipy 의 stats 모듈 내에 존재
    * 전체에서 특정 비율에 해당하는 z 점수를 구할 때 사용
    * ppf : 백분위수 함수 (Percent point function)
  - norm.pdf() 메서드
    * scipy 의 stats 모듈 내에 존재
    * pdf : 확률 밀도 함수 (probability density function)
  - norm.pmf() 메서드
    * scipy 의 stats 모듈 내에 존재
    * pmf : 확률 질량 함수 (probability mass function)

* 중심극한정리
  - 무작위로 샘플을 뽑아 만든 표본의 평균은 정규분포에 가깝다는 이론
  - sample() 메서드
    * 무작위 샘플링을 위해 사용

* 신뢰구간
  - 표본의 파라미터(평균 등)가 속할 것이라 믿는 모집단의 파라미터 범위
  - 표본의 평균만 알 때 z 점수 공식
  $$z=\frac{x-\mu}{s} = \frac{x-\mu}{\sigma/\sqrt{n}}$$
  $$\mu=x-z\frac{\sigma}{\sqrt{n}}$$

* 가설 검정
  - 표본에 대한 정보를 사용해 모집단의 파라미터에 대한 가정을 검정하는 것
  - 가설 분류
    * 영가설(귀무가설) : 표본 사이에 통계적으로 의미가 없다고 예상되는 가설
    $$H_0 : \mu_{python} - \mu_{c^{++}} = 0$$
    * 대립가설 : 표본 사이에 통계적인 차이가 있다는 가설
    $$H_A : \mu_{python}-\mu_{C^{++}} \ne 0$$
  - 두 모집단의 평균에 대한 Z 점수 공식
  $$ z = \frac{(x_1-x_2)-(\mu_1-\mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}} $$
  - t 검정으로 가설 검증하기
    * t-분포
      - 정규분포와 비슷하지만, 중앙은 조금 더 낮고 꼬리가 더 두꺼운 분포
      - 표본의 크기가 30 이하일 때 t-분포를 사용하는 것이 좋다,
    * ttest_ind() 함수
      - 두 표본의 평균을 비교
      - t-분포인 두 표본을 비교하는 t-검정을 수행
      - t 점수와 p-값을 반환
    
* 순열 검정
  - 모집단의 분포가 정규분포를 따르지 않거나 모집단의 분포를 알 수 없을 때 가설 검증
  - 모집단의 파라미터를 추정하지 않기 때문에 비모수검정 방법 중 하나
  - 순서
    * 1. 두 표본의 평균 차이를 계산
    * 2. 두 표본을 섞고 다시 무작위로 원래의 표본의 크기와 동일하게 나눈다.
    * 3. 나눈 그룹에서 다시 평균의 차이를 계산
    * 4. 2, 3 번 과정을 반복하여 p-값을 계산

## 7-2. 머신러닝으로 예측하기
* 머신러닝 용어
  - 인공지능 > 머신러닝 > 지도학습 + 비지도학습
  - 사이킷런
    * 파이썬의 대표적인 머신러닝 패키지
  - 모델
    * 추상적으로 선형 회귀라는 특정 알고리즘을 대변
    * 머신러닝으로 학습된 패턴을 저장하는 소프트웨어 객체를 의미
  - 지도학습 
    * 데이터에 있는 각 샘플에 대한 정답(타깃)을 알고 있는 경우 사용
    * 입력 : 타깃을 맞추기 위해 모델이 재료로 사용하는 데이터
  - 비지도 학습
    * 입력 데이터는 있지만 타깃(정답)이 없는 경우 사용
    * 머신러닝으로 의미있는 패턴을 찾는다.
    * 예) 군집 분석

* 모델 훈련하기
  - 데이터를 훈련 세트와 테스트 세트로 분류
  - 테스트 데이터는 전체 데이터의 20-25% 정도를 사용
  - train)test_split() 함수
    * 사이킷런의 model_selection내에 존재
    * 입력된 데이터를 무작위로 섞은 후, 75%를 훈련세트로, 25%를 테스트 세트로 나눈다.
  - 일반적으로 행이 샘플, 열을 특성이라 한다.
  - LinearRegression() 클래스
    * 사이킷런의 linear_model 모듈내에 존재
    * 선형 회귀 알고리즘
    * 선형 회귀 알고리즘의 객체를 반환
    * 객체의 fit() 메서드를 호출하여 모델 훈련
    * 속성
      - coef_ : 기울기가 저장됨
      - intercept_ : 절편이 저장됨

* 모델 평가하기 : 결정계수
  - 결정계수
  $$R^2 = 1 - \frac{{(타깃 - 예측)}^2}{{(타깃 - 평균)}^2}$$
  - score() 메서드
    * 머신러닝 모델이 얼마나 유용한지 평가
    * 0-1사이의 값을 가지며, 1에 가까울수록 정확하다는 의미
    * 결정계수
  - to_frame() 메서드
    * 시리즈 -> 데이터프레임 형태로 변환

* 선형 회귀 : 연속적인 값 예측하기
  - 선형 함수를 사용해 모델을 만드는 알고리즘
  $$y=a \times x + b$$

* 로지스틱 회귀 : 카테고리 예측하기
  - 회귀 : 타깃이 연속적인 실수인 경우 사용하는 모델
  - 분류 : 타깃이 연속적인 값이 아닌, 정수와 같이 떨어져 있는 수인 경우 사용하는 모델
    * 이진 분류 : 두 개의 카테고리로 구분
    * 다중 분류 : 세 개 이상의 카테고리로 구분
  - 분류 알고리즘 용어
    * 클래스 : 타깃이 되는 카테고리
    * 이진 분류의 경우 타깃은 0(음성 클래스), 1(양성 클래스)
  - 로지스틱 회귀 : 분류 알고리즘의 대표적인 알고리즘
    * 로지스틱 함수를 사용해 연속적인 실수 출력값을 1 또는 0으로 변환
    $$z = a \times x + b$$
    $$y = \frac{1}{1+e^{-z}}$$
    * 실수 z를 로지스틱 함수에 통과하면 0~1 사이의 y를 얻는데, 0.5 이하인 경우 음성 클래스(0), 0.5 이상인 경우 양성 클래스(1)로 예측
  - LogisticRegression() 클래스
    * 로지스틱 회귀 모델
    * 사이킷런의 linear_model 내에 존재
  - DummyClassifier() 모델
    * 가장 많은 클래스로 무조건 예측을 수행하는 더미 모델
    * 분류 일 경우 사용
  - DummyRegressor() 모델
    * 회귀 일 경우 사용
    * 무조건 타깃의 평균을 예측
