# 1. 데이터 분석을 시작하며

## 1-1. 데이터 분석이란
* 용어 정리
  - 데이터 분석
    * 의사 결정을 돕기 위해 데이터를 조사, 정제, 변환, 모델링 하는 과정
    * 올바른 의사 결정을 돕기 위한 통찰을 제공
  - 데이터 과학
    * 통계학, 데이터 분석, 머신러닝, 데이터 마이닝 등을 아우르는 큰 개념
    * 문제 해결을 위한 최선의 솔루션을 만드는 데 초점을 맞춘다.
  - 데이터 마이닝
    * 데이터에서 패턴 혹은 지식을 추출하는 작업
  - 머신러닝
    * 데이터에서 자동으로 규칙을 학습하여 문제를 해결하는 소프트웨어

* 데이터 분석
  - 통계적 관점에서의 데이터 분석 (좁은 의미)
    1) 기술 통계 : 관측이나 실험을 통해 수집한 데이터를 정량화 하거나 요약하는 기법
    2) 탐색적 데이터 분석 : 데이터를 시각적으로 표현하여 주요 특징을 찾고 분석하는 방법
    3) 가설 검정 : 주어진 데이터를 기반으로 특정 가정이 합당한지 평가하는 통계 방법
  - 넓은 의미에서의 데이터 분석(+ 통계적 관점)
    1) 데이터 수집
    2) 데이터 처리
    3) 데이터 정제
    4) 모델링

* 데이터 분석을 위한 프로그래밍 언어
  - 파이썬
  - R
  - SQL(데이터가 데이터 베이스 형태일 경우 사용 가능)

* 파이썬 필수 패키지
  - 넘파이 (numpy)
    * 고성능 과학 계산과 다차원 배열을 위한 파이썬 패키지
  - 판다스(pandas)
    * 파이썬 데이터 분석을 위한 전문 패키지
    * 엑셀 시트처럼 숫자와 문자를 섞어서 표 형태로 저장할 수 있는 데이터프레임을 사용
  - 맷플롯립(matplotlib)
    * 파이썬 데이터 시각화를 위한 기본 패키지
    * 정적 혹은 동적인 그래프를 그릴 수 있고, 고품질의 다양한 그래프를 제공
  - 사이파이(scipy)
    * 넘파이를 기반으로 구축된 수학과 과학 계산 전문 패키지
    * 미분, 적분, 확률, 선형대수, 최적화 등을 알고리즘을 ㅗ구현
  - 사이킷런(scikit-learn)
    * 파이썬의 독보적인 머신러닝 패키지
    * 넘파이와 사이파이에 크게 의존

## 1-3. 데이터 찾기
* 공개 데이터 세트 대표 사이트와 유명 포럼
  - 국내
    * 공공 데이터 포털
    * 통합 데이터 지도
    * AI 허브
    * 국가 통계 포털
  - 해외
    * 구글 데이터 세트 검색
    * 캐글 데이터 세트
    * 위키피디아 머신러닝 데이터 세트
    * 아마존 웹서비스 오픈 데이터
    * UCI 머신러닝 데이터 저장소
  - 온라인 포럼
    * 데이터 분석 커뮤니티
    * 캐글 코리아
    * 텐서플로 코리아
    * 파이토치 코리아
    * 사이킷런 코리아

* 코랩에서 데이터 확인
  - CSV file
    * 콤마(,)로 구분된 텍스트 파일
    * 한 줄이 하나의 행(레코드, record)이며 레코드는 콤마로 구분된 여러 열(필드, field)로 구성
    * 데이터는 엑셀처럼 표 형태여야 하기 때문에 레코드에 있는 필드 개수는 모두 동일
    * 행은 한 줄로 표현되고, 열은 콤마로 구분
  
* gdown 패키지
  - 구글 드라이브에서 코랩으로 다운로드 할 때 사용

* CSV 파일 출력
  - 텍스트 파일 이므로, 파이썬의 open() 함수로 읽을 수 있다.
    * open() 함수는 기본적으로 텍스트 파일이 UTF-8 형식으로 저장되어 있다고 가정
    * 매개변수
      - mode : 바이너리 읽기 모드인 'rb' 로 지정하면 문자 인코딩 형식에 상관없이 파일을 열 수 있다.
      - encoding : 이용하여 인코딩 방식 지정 가능
  - chardet.detect() : 문자의 인코딩 방식을 알아낼 수 있다.

* 데이터 프레임 다루기 : 판다스
  - csv 파일을 읽어 데이터 프레임이라는 표 형식 데이터로 저장
    * 표 형식 데이터 : 행과 열로 구성된 데이터 구조
  - 시리즈
    * 1차원 배열과 비슷
    * 시리즈에 담긴 데이터는 동일한 형식이여야 함
  - 데이터 프레임
    * 데이터를 가로 세로로 나열한 2차원 배열과 유사
    * 열마다 다른 데이터 타입을 사용 가능 (같은 열 내 데이터는 같은 종류)
  - head() : 데이터 프레임의 처음 다섯 개 행 확인 가능
    * 첫번째 열은 데이터 프레임의 인덱스가 출력됨
    * 첫번째 행은 열 이름으로 인식
    * 첫번째 행(열 제목)이 중복되는 경우 자동으로 이름 뒤에 숫자를 붙여 고유한 이름을 만듬
  - read_csv() : csv 파일을 읽을 때 사용
    * csv파일을 읽을 때 같은 열에 어떤 종류의 데이터가 저장되어 있는 지 자동으로 파악함
    * 매개변수
      - encoding : 이용하여 인코딩 방식 지정 가능
      - low_memmory : True는 파일을 나누어 읽는 것이고, False는 파일을 한번에 읽는 것
      - dtype : 데이터 타입을 지정하면 데이터 타입에 대한 문제의 소지가 없어짐.
      - header : 첫 행이 열 이름이 아닐 경우 None으로 지정하여 열 이름이 아닌 것을 표시
      - names : 열 이름 리스트를 따로 전달해 줄 때 사용
      - index_col : 이미 csv 파일에 인덱스가 있는 경우 사용
  - to_csv() : csv 파일로 저장할 때 사용
    * 기본적으로 UTF-8 형식으로 저장
    * 매개변수는
      - index : 인덱스를 빼고 저장하고자 할 때 사용
