1. 데이터 프레임 다루기 : 판다스

```python
import pandas as pd
# 기본 : 자동으로 파악한 데이터 타입이 달라지면 경고 발생
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR')
df.head()
```

```python
# 파일을 한번에 읽기(메모리 고려 X)
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR', low_memory='False')
df.head()
```

```python
# 파일을 나누어 읽을 경우 자동으로 파악한 데이터 타입이 달라지면 경고 발생됨
# 나누어 읽기 + 경고 발생되지 않게 하기 위해 데이터 타입을 지정해줌
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR', 
                 dtype={'ISBN': str, '세트 ISBN': str, '주제분류번호': str})
df.head()
```

```python
# 첫 행이 열 이름이 아닌 경우
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR', 
                 dtype={'ISBN': str, '세트 ISBN': str, '주제분류번호': str},
                 header=None, names=['번호','도서명','저자','출판사','발행년도','ISBN','세트 ISBN','부가기호','권','주제분류번호','도서권수','대출건수','등록일자'])
df.head()
```

```python
# 데이터 프레임을 csv 파일로 저장
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR', 
                 dtype={'ISBN': str, '세트 ISBN': str, '주제분류번호': str})
df.to_csv('ns_202104.csv')
```
