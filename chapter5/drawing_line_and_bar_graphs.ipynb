```python
# 파일 download
import gdown
gdown.download("http://bit.ly/3pK7iuu", 'ns_book7.csv', quiet=False)
```

```python
# 데이터프레임 로드
import pandas as pd
ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
# 연도별 도서 개수 구하기
count_by_year = ns_book7['발행년도'].value_counts()
count_by_year
```

```python
# 시간 순서대로 정렬
count_by_year = count_by_year.sort_index()
count_by_year
```

```python
# index 속성이 2030년보다 작거나 같은 데이터만 추출
count_by_year = count_by_year[count_by_year.index <= 2030]
count_by_year
```

```python
# 주제별 도서 개수 구하기
import numpy as np

def kdc_1st_char(no):
  if no is np.nan:
    return '-1'
  else:
    return no[0]

count_by_subject = ns_book7['주제분류번호'].apply(kdc_1st_char).value_counts()
count_by_subject
```

```python
# 선 그래프 그리기를 위해 기본 설정 하기
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100
```

```python
# count_by_year 그리기
plt.plot(count_by_year.index, count_by_year.values)
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of bppls')
plt.show()
```

```python
# 선 모양과 색상 바꾸기
plt.plot(count_by_year, marker='.', linestyle=':', color='red')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
# 선모양과 색상을 한번에 바꾸기
plt.plot(count_by_year, '-g*')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.show()
```

```python
# 선 그래프 눈금 개수 조절 및 마커에 텍스트 표시하기
plt.plot(count_by_year, 'g-*')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947, 2030, 10))

for idx, val in count_by_year[::5].items():
  plt.annotate(val, (idx, val))
plt.show()
```

```python
# 텍스트 위치 변경
plt.plot(count_by_year, '-*g')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947, 2030, 10))

for idx, val in count_by_year[::5].items():
  plt.annotate(val, (idx, val), xytext = (idx+1, val+10))
plt.show()
```

```python
plt.plot(count_by_year, '-g*')
plt.title('Books by year')
plt.xlabel('year')
plt.ylabel('number of books')
plt.xticks(range(1947, 2030, 10))
for idx, val in count_by_year[::5].items():
  plt.annotate(val, (idx, val), xytext=(2,2), textcoords='offset points')
plt.show()
```

```python
# 막대 그래프 그리기
plt.bar(count_by_subject.index, count_by_subject.values)
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')
for idx, val in count_by_subject.items():
  plt.annotate(val, (idx, val), xytext=(0,2), textcoords='offset points')
plt.show()
```

```python
# 텍스트 정렬, 막대 조질 및 색상 바꾸기
plt.bar(count_by_subject.index, count_by_subject.values, color='blue', width=0.7)
plt.title('Books by subject')
plt.xlabel('subject')
plt.ylabel('number of books')
for idx, val in count_by_subject.items():
  plt.annotate(val, (idx, val), color='green', ha='center', xytext=(0,2), textcoords='offset points', fontsize=8)
plt.show()
```

```python
# 가로 막대 그래프 그리기
plt.barh(count_by_subject.index, count_by_subject.values, color='blue', height=0.7)
plt.title('Books by subject')
plt.xlabel('number of books')
plt.ylabel('subject')
for idx, val in count_by_subject.items():
  plt.annotate(val, (val, idx), color='green', va='center', xytext=(2,0), textcoords='offset points', fontsize=8)
plt.show()
```

```python
# 그래프를 이미지로 저장하기
plt.barh(count_by_subject.index, count_by_subject.values, height=0.7, color='blue')
plt.title('Books by subject')
plt.xlabel('number of books')
plt.ylabel('subject')
for idx, val in count_by_subject.items():
  plt.annotate(val, (val, idx), xytext=(2,0), textcoords='offset points', va='center', fontsize=8, color='green')
plt.savefig('books_by_subject.png')
plt.show()
```

```python
# 저장한 이미지를 불러와 출력
img = plt.imread('books_by_subject.png')
plt.figure(figsize=(8,6))
plt.imshow(img)
plt.axis('off')
plt.show()
```
