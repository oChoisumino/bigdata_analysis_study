```python
# 이미지 읽기
import matplotlib.pyplot as plt
img = plt.imread('jupiter.png')
img.shape
```

```python
# 이미지 출력하기
plt.imshow(img)
plt.show()
```

```python
# 축이 표시되지 않도록 이미지 출력
plt.figure(figsize=(8,6))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# 이미지 저장하기
plt.imsave('jupiter.jpg', img)
```
