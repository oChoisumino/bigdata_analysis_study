# 2. 데이터 수집하기

## 2-1. API 사용하기
* API란
  - 두 프로그램이 서로 대화하기 위한 방법을 정의한 것

* HTTP
  - 인터넷에서 웹 페이지를 전송하는 기본 통신 방법
  - 웹 페이지는 서비스하기 위해 웹 서버 소프트웨어 사용
  - 웹 서버 프로그램은 웹 브라우저와 통신할 때 HTTP 라는 프로토콜 사용
    * 웹 서버 프로그램 : 엔진엑스, 아파치 등
  - 웹 브라우저가 웹 서버에 웹페이지를 요청하고, 웹 서버는 요청에 맞는 웹 페이지를 웹 브라우저에 전송
    * 웹 서버와 웹 브라우저는 소프트웨어 프로그램이다.

* HTML
  - 웹 페이지 문서
  - 웹 브라우저가 화면에 표시할 수 있는 문서의 한 종류이자 웹 페이지를 위한 표준 언어
  - 웹 서버와 웹 브라우저는 HTTP 프로토콜을 사용해 웹 페이지를 구성하는 HTML 데이터를 안정적으로 주고받는다.
  - 웹 기반 API는 HTTP 프로토콜을 사용하지만 HTML을 주고받지 않고 CSV, JSON, XML 등을 주고 받는다.
    * HTML의 구조가 복잡하기 때문

* 파이썬에서 JSON 데이터 다루기
  - 자바스크립트 언어를 위해 만들어졌지만, 현재는 범용적인 포맷으로 사용
  - JSON
    * 딕셔너리와 리스트를 중첩해 놓은 것과 비슷
    * 중괄호를 사용하고, 키와 값을 콜론으로 연결
    * {"name":"혼자 공부하는 데이터 분석"} => 키 : name, 값 : 혼자 공부하는 데이터 분석
    * json 객체를 대괄호 안에 나열하고, 콤마로 구분하면 json 객체 배열로 표현 가능
  - json.dumps()
    * 딕셔너리를 JSON 형식에 맞는 문자열로 변경
    * json 패키지 사용
    * 매개변수
      - ensure_ascii : 한글이 포함된 경우, False를 입력해주어야 한다.
  - json.loads()
    * JSON 문자열을 파이썬 객체로 변환
  - pd.read_json()
    * JSON 문자열을 데이터프레임으로 변환

* 파이썬에서 XML 데이터 다루기
  - XML 
    * 컴퓨터와 사람이 모두 읽고 쓰기 편한 문서 포맷을 위해 고안됨
    * element들이 계층 구조를 이루면서 정보를 표현
    * 각 element는 시작 태그와 종료 태그로 감싼다.
    * 태그는 <>로 감싸고, 태그 이름은 영문자와 숫자를 사용
    * 시작태그(<이름>)와 종료태그(</ 이름>)의 이름은 같아야 한다.
  - xml.etree.ElementTree.fromstring()
    * ElementTree 모듈 아래에 정의된 Element 클래스의 객체를 반환
  - xml.etree.ElementTree.findtext()
    * 자식 element를 탐색하여 자동으로 텍스트를 반환
  - xml.etree.ElementTree.findall()
    * 동일한 이름을 가진 여러 개의 자식 element를 찾을 때 사용

* API로 데이터 찾기 
  - API를 호출하는 URL 작성하기 (도서관 정보나루 공개 API 사용)
    * 공개 API를 사용하려면 호출 URL이 필요
    * 호출 URL : http://data4library.kr/api/loanItemSrch
    * 파라미터
      - format : 지정하지 않으면 XML 문서로 반환
      - startDt : 검색 시작 일자
      - endDt : 검색 종료 일자
      - age : 연령대
      - authKey : 인증키, 정보나루 사이트에서 인증키를 신청하여 받아야 한다.
    * 조회할 값은 호출 URL 뒤에 파라미터를 뒤에 연결 -> HTTP GET 방식
      - 파라미터와 값은 '=' 문자로 연결
      - 파라미터 사이는 '&' 문자로 연결
      - 호출 URL과 파라미터는 '?' 문자로 연결
  - requests 패키지
    * 파이썬에서 URL을 호출하여 데이터를 받을 수 있는 방법 중 하나
    * get() 함수
      - API 호출의 결과를 담고 있는 requests 패키지읲 Response 클래스 객체를 반환
    * json() 메서드 
      - 웹서버로부터 받은 JSON 문자열을 파이썬 객체로 변환하여 반환

* 공개 API 로 웹에서 데이터 가져오기
  - requests 패키지를 활용하여 필요한 공개 API를 찾아서 파이썬 프로그램에서 호출 & 데이터프레임으로 변환
  - get() & json()

## 2-2. 웹 스크래핑 사용하기
* 웹 스크래핑, 웹 크롤링
  - 프로그램으로 웹사이트의 페이지를 옮겨가면서 데이터를 추출하는 작업

* DataFame 사용
  - json file 을 읽어 DataFrame 으로 변환
    * data = pd.read_json('파일명')
  - 데이터프레임에서 일부 열만 선택하여 새로운 데이터프레임 형성
    * new_data = 데이터프레임[["선택열1", "선택열2", ...]]
  - 데이터프레임에서 행과 열 선택하기
    * loc 메서드 이용
      - 대괄호를 사용하여 행의 목록과 열의 목록을 받음
      - 예1) 데이터프레임.loc[["선택행1", "선택행2", ...], ["선택열1", "선택열2", ...]]
      - 예2) 데이터프레임.loc["시작행":"끝행", "시작열":"끝열"]

* 검색 결과 페이지 HTML 가져오기
  - requests.get() 함수 이용

* HTML에서 데이터 추출하기(뷰티플수프)
  - HTML 안에 있는 내용을 찾을 때 사용
  - 크롬 개발자 도구로 HTML 태그 찾기
    * 원하는 데이터가 HTML 어디에 위치해 있는 지 알아야 한다.
    * 크롬에서 마우스 오른쪽 버튼을 클릭하여 [검사]를 클릭하여 개발자 도구 창 열기
    * 개발자 도구 창 메뉴 바에서 [Select] 아이콘을 클릭하고, 마우스 커서를 올려놓으면 HTML 위치가 나타남
  - BeautifulSoup 클래스 생성
    * BeautifulSoup 패키지에서 BeautifulSoup 클래스 import
    * 객체 생성
      - 1번째 매개변수 : 파싱할 HTML 문서
      - 2번째 매개변수 : 파싱에 사용할 파서
  - find() 메서드
    * 태그 위치를 찾을 때 사용
    * 첫번째 매개변수 : 찾을 태그 이름 지정
    * attrs 매개변수 : 찾으려는 태그의 속성을 딕셔너리로 지정
  - find_all() 메서드
    * 테이블 태그를 리스트로 가져올 때 사용
    * 특정 HTML 태그를 모두 찾아서 리스트로 반환
    * 매개변수 : 찾을 태그 이름 지정
  - get_text() 메서드
    * 태그 안의 텍스트 가져올 때 사용

* 전체 도서의 쪽수 구하기
  1) 온라인 서점의 검색 결과 페이지 URL을 만든다.
  2) requests.get() 함수로 검색 결과 페이지의 HTML을 가져온다.
  3) 뷰티플수프로 HTML을 파싱한다.
  4) 뷰티플수프의 find() 메서드로 <a> 태그를 찾아 상세 페이지  URL을 추출한다.
  5) requests.get() 함수로 다시 도서 상세 페이지의 HTML을 가져온다.
  6) 뷰티플수프로 HTML을 파싱한다.
  7) 뷰티플수프의 find() 메서드로 '품목정보' <div> 태그를 찾는다.
  8) 뷰티플수프의 find_all() 메서드로 '쪽수'가 들어있는 <tr> 태그를 찾는다.
  9) 앞에서 찾은 테이블의 행에서 get_text() 메서드로 <td> 태그에 들어있는 쪽수를 가져온다.

* 데이터프레임 행 혹은 열에 함수 적용
  - apply() 메서드
    * 첫번째 매개변수 : 실행할 함수
    * axis : 0은 각 열에 대해 함수를 적용하고, 1은 각 행에 대해 함수를 적용한다.

* 데이터프레임과 시리즈 합치기
  - merge() 함수
    * 매개변수
      - 첫 번쨰와 두 번째 매개변수 : 합칠 데이터프레임이나 시리즈 객체
      - left_index : 왼쪽 인덱스를 기준으로 합치기
      - right_index : 오른쪽 인덱스를 기준으로 합치기
      - on : 합칠 때 기준이 되는 열 지정
        * 이 열은 두 데이터프레임에 모두 존재해야 한다. 다를 경우 합쳐지지 않는다.
      - how : 합쳐질 방식 지정
        * 기본값 : inner로, how 매개변수를 생략하면 기본적으로 두 데이터의 임의 값이 같은 행만 합친다.
        * left : 첫번째 데이터프레임을 기준으로 두번째 데이터프레임을 합친다.
        * right : 두번째 데이터프레임을 기준으로 첫 번째 데이터프레임을 합친다.
        * outer : 두 데이터프레임의 모든 행을 유지하면서 합친다.
      - left_on, right_on : 합칠 기준이 되는 열 이름이 서로 다를 경우, 각기 지정 가능
      - left_index, right_index : 합칠 기준이 열이 아닌 인덱스일 경우 각각 지정 가능

* 웹 스크래핑 할 때 주의점
  - 웹사이트에서 스크래핑을 허락했는지 확인
  - HTML 태그를 특정할 수 있는 지 확인
    * 태그 이름이나 속성 등 필요한 HTML 태그를 특정할 수 있어야 가능
