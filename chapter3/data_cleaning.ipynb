# 3. 데이터 정제하기

## 3-1. 불필요한 데이터 삭제하기
* 데이터 정제
  - 데이터에서 손상되거나 부정확한 부분을 수정하고, 불필요한 데이터를 삭제하거나 불완전한 값을 교체하는 등의 작업
  - 데이터를 분석 목적에 맞게 변환하는 데이터 랭글링(data wrangling), 데이터 먼징(data munging)의 일부로 수행

* 열 삭제하기
  - case 1. loc 메서드에 슬라이싱 사용
    - 데이터프레임.loc[행,열]
  - case 2. loc 메서드와 불리언 배열
    - 사용할 열을 불리언 배열로 나타내어 적용
    - 데이터프레임.columns != '열 이름' 을 이용하여 불리언 배열을 얻을 수 있다.
      * 반환된 결과는 넘파이 배열
    - loc 메서드와 불리언 배열을 이용하면, 배열에 True로 표시된 열만 선택할 수 있다.
  - case 3. drop() 메서드
    - 데이터프레임의 행이나 열을 삭제
    - 매개변수 
      * 첫 번째 매개변수에 삭제하려는 이름을 전달 (여러 개일 경우 리스트로 전달)
      * axis 매개변수로 열 또는 행을 지정
        - axis 의 default는 0으로 선택한 행을 삭제하고, 1은 선택한 열을 삭제
      * inplace 매개변수로 현재 데이터 프레임을 바로 수정함
  - case 4. dropna() 메서드
    - 비어있는 값인 NaN이 하나 이상 포함된 행이나 열을 삭제
    - 매개변수
      * axis 매개변수로 열 또는 행 지정
        - axis 의 default는 0으로 NaN인 행을 삭제하고, 1은 NaN인 열을 삭제
      * how 매개변수로 방식 지정 가능
        - 'all' 인 경우 모든 값이 NaN인 열을 삭제한다.

* 행 삭제하기
  1) drop() 메서드 
  2) [ ] 연산자와 슬라이싱
    - 행을 선택할 때 자주 사용
    - [ ] 연산자에 슬라이싱을 전달하여 행을 선택
  3) [ ] 연산자와 불리언 배열
    - 행을 선택할 때 자주 사용
    - 비교 연산자를 활용해 원하는 행은 True로 표시하고 제외할 행은 False로 표시한 불리언 배열을 만들어 사용

* 중복된 행 찾기
  - duplicated() 메서드를 사용
    * 중복된 행 중에서 처음 행을 제외한 나머지 행은 True로, 그 외에 중복되지 않은 나머지 모든 행은 False로 표시
    * 매개변수
      - subset : 일부 열을 기준으로 중복된 행을 찾을 때 사용
      - keep : False일 경우, 중복된 행을 모두 True로 표시한 불리언 배열을 반환

* 그룹별로 모으기
  - groupby() 메서드 사용
    * 매개변수
      - by : 행을 합칠 때 기준이 되는 열 지정
      - dropna : False일 경우, NaN 이 있는 행을 삭제하지 않는다.

* 원본 데이터프레임 인덱스 설정하기
  - set_index() 메서드 사용
    * 매개변수
      - 인덱스 이름 리스트를 매개변수로 전달
      - inplace : True일 경우 새로운 데이터프레임을 반환하지 않고, 현재 데이터프레임을 수정한다.

* 업데이트 하기
  - update() 메서드 사용
    * 다른 데이터프레임을 사용해 원본 데이터프레임의 값을 업데이트 할 때 사용

* 인덱스 열 해제
  - reset_index() 메서드 사용
    * 인덱스 열을 해제
    * 데이터프레임 인덱스를 재설정할 수 있다.

## 3-2. 잘못된 데이터 수정하기
* 누락된 값 처리하기
  - 판다스는 누락된 값을 기본적으로 'NaN'으로 표시
  - isna() 메서드
    * 누락된 값 확인
    * 각 행이 비어있는지를 나타내는 불리언 배열 반환
  - 누락된 값으로 표시하기
    * 데이터가 정수 인 경우는 None 입력
    * 데이터가 문자열인 경우 np.nan 입력

* astype() 메서드
  - 데이터 타입 지정할 때 사용

* 누락된 값 바꾸기
  - loc() 메서드
    * 누락된 값을 가리키는 불리언 배열(isna())과 loc 메서드를 사용하여 누락된 값을 원하는 값으로 변경
  - fillna() 메서드
    * NaN인 값을 원하는 값으로 변경할 때 사용
    * 데이터프레임.fillna(값) : 원하는 값을 전달하면 NaN을 대체
    * 데이터프레임[열] : 특정 열의 값만 변경
    * 데이터프레임.fillna(딕셔너리) : 선택한 열을 원하는 값으로 변경
  - replace() 메서드
    * 특정 값을 원하는 값으로 변경할 때 사용
    * case 1. 바꾸려는 값이 하나일 때
      - replace(원래 값, 새로운 값)
    * case 2. 바꾸려는 값이 여러 개일 때
      - replace([원래 값1, 원래 값2], [새로운 값1, 새로운 값2])
      - replace({원래 값1:새로운 값1, 원래 값2:새로운 값2})
    * case 3. 열마다 다른 값으로 바꿀 때
      - replace({열 이름:원래 값}, 새로운 값)
      - replace({열 이름 : {원래 값:새로운 값}})

* 정규 표현식 (정규식)
  - 문자열 패턴을 찾아서 대체하기 위한 규칙 모음
  - replace() 메서드를 이용하고, regex 매개변수 옵션을 True로 지정
  - 숫자 찾기 (\d)
    * 정규식에서 숫자를 나타내는 기호는 \d
    * 네자리 연도의 경우는 \d\d\d\d
    * 그룹으로 묶을 경우 괄호 ()를 사용
    * 패턴 안에 있는 그룹을 나타낼 때는 \1, \2처럼 사용하고, 그룹의 번호는 패턴 안에 등장하는 순서대로 매겨진다.
  - 문자 찾기 : 마침표(.)
    * 마침표(.)는 어떤 문자에도 대응하는 정규 표현식 문자
      - 글자수를 모르는 경우 *와 함께 사용 (.*)
    * 괄호를 사용할 때는 역슬래시(\)를 앞에 붙여서 사용
    * 공백 문자를 표현할 때는 \s 정규 표현식 사용

* 문자열 처리 함수 제공
  - 판다스 시리즈 객체는 str 속성 아래에 다양한 문자열 처리 함수를 제공
  - contains('값') : '값' 문자열 패턴을 포함하고 있는 지 검사
    * 숫자에 대응하는 정규 표현식 : \d
    * 모든 문자에 대응하는 표현 : \D
    * 매개변수
      - na : True일 경우 누락된 행을 True로 표시
